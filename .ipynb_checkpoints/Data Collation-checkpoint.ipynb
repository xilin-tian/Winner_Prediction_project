## Data Collation

The purpose of this document is to process the different datasets collected from the website. We obtained the records of the teams that made it to the Worlds for the entire year as well as other tournament data from [oracleselixir](https://oracleselixir.com/), and then from [League_of_Legends_Esports_Wiki](https://lol.fandom.com/wiki/League_of_Legends_Esports_Wiki) to get the wins and losses for each game of the Worlds each year, starting with the main tournament. In total, data was collected for a total of 5 years from 2019-2023.

A total of 16 teams compete each year. In order to concatenate the two datasets together, we decided to subtract the data of the red team from the data of the blue team for each game to get the difference between the two teams as the INDEPENDENT variable of the dataset; and whether the blue team wins or not as the DEPENDENT variable of the dataset. thus we obtain a new dataset that will be based on the predicted the winning percentage of the blue team for further analysis.

```python
import pandas as pd
```

```python
# Load the Data gather from Worlds Team for each year.   source: https://oracleselixir.com/stats/teams/byTournament
# Prearranged: Column WR is calculate by GP(Game Played) divided by W(Win)
year2023StatForWorldsTeam = pd.read_excel(r"year2023StatForWorldsTeam.xlsx") 
year2022StatForWorldsTeam = pd.read_excel(r"year2022StatForWorldsTeam.xlsx") 
year2021StatForWorldsTeam = pd.read_excel(r"year2021StatForWorldsTeam.xlsx") 
year2020StatForWorldsTeam = pd.read_excel(r"year2020StatForWorldsTeam.xlsx") 
year2019StatForWorldsTeam = pd.read_excel(r"year2019StatForWorldsTeam.xlsx") 

# Load the Win/Loss for each team in Worlds.   source: https://lol.fandom.com/wiki/League_of_Legends_Esports_Wiki
WinLose23 = pd.read_excel(r"2023WinLose.xlsx")
WinLose22 = pd.read_excel(r"2022WinLose.xlsx")
WinLose21 = pd.read_excel(r"2021WinLose.xlsx")
WinLose20 = pd.read_excel(r"2020WinLose.xlsx")
WinLose19 = pd.read_excel(r"2019WinLose.xlsx")
```

```python
Team1vsTeam2 = pd.DataFrame(columns = list(year2023StatForWorldsTeam.columns.values)) # Create new dataset for the secondary analysis
Team1vsTeam2 = Team1vsTeam2.drop(columns=['team'])

# Calculate the difference in data between the teams on the blue side（Team1） and the red side（Team2）
for i in range(len(WinLose23)):
    team1_23 = WinLose23.loc[i][0] # Find the name for Team1
    team2_23 = WinLose23.loc[i][1] # Find the name for Team2
    team1data_23 = year2023StatForWorldsTeam.loc[year2023StatForWorldsTeam["team"] == team1_23].set_index('team') # Get the team1 data by the name we extract
    team2data_23 = year2023StatForWorldsTeam.loc[year2023StatForWorldsTeam["team"] == team2_23].set_index('team') # Get the team2 data by the name we extract
    compdata_23 = team1data_23.sub(team2data_23.squeeze())# Let team1 data minus team2 data, find the gap between 2 teams
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata_23]) # Add to our Analysis dataset
    WinLose_23 = WinLose23['IsTeam1Won']
    
# Repeat the processing for the previous year
for i in range(len(WinLose22)): 
    team1_22 = WinLose22.loc[i][0] 
    team2_22 = WinLose22.loc[i][1] 
    team1data_22 = year2022StatForWorldsTeam.loc[year2022StatForWorldsTeam["team"] == team1_22].set_index('team') 
    team2data_22 = year2022StatForWorldsTeam.loc[year2022StatForWorldsTeam["team"] == team2_22].set_index('team')
    compdata_22 = team1data_22.sub(team2data_22.squeeze()) 
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata_22])
    WinLose_22to23 = pd.concat([WinLose_23, WinLose22['IsTeam1Won']])
    
for i in range(len(WinLose21)):
    team1_21 = WinLose21.loc[i][0]
    team2_21 = WinLose21.loc[i][1] 
    team1data_21 = year2021StatForWorldsTeam.loc[year2021StatForWorldsTeam["team"] == team1_21].set_index('team') 
    team2data_21 = year2021StatForWorldsTeam.loc[year2021StatForWorldsTeam["team"] == team2_21].set_index('team') 
    compdata_21 = team1data_21.sub(team2data_21.squeeze())
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata_21])
    WinLose_21to23 = pd.concat([WinLose_22to23, WinLose21['IsTeam1Won']])
    
for i in range(len(WinLose20)):
    team1_20 = WinLose20.loc[i][0] 
    team2_20 = WinLose20.loc[i][1]
    team1data_20 = year2020StatForWorldsTeam.loc[year2020StatForWorldsTeam["team"] == team1_20].set_index('team')
    team2data_20 = year2020StatForWorldsTeam.loc[year2020StatForWorldsTeam["team"] == team2_20].set_index('team')
    compdata_20 = team1data_20.sub(team2data_20.squeeze()) 
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata_20])
    WinLose_20to23 = pd.concat([WinLose_21to23, WinLose20['IsTeam1Won']])
    
for i in range(len(WinLose19)):
    team1_19 = WinLose19.loc[i][0]
    team2_19 = WinLose19.loc[i][1]
    team1data_19 = year2019StatForWorldsTeam.loc[year2019StatForWorldsTeam["team"] == team1_19].set_index('team') 
    team2data_19 = year2019StatForWorldsTeam.loc[year2019StatForWorldsTeam["team"] == team2_19].set_index('team')
    compdata_19 = team1data_19.sub(team2data_19.squeeze())
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata_19])
    WinLose_19to23 = pd.concat([WinLose_20to23, WinLose19['IsTeam1Won']])
    
Team1vsTeam2 = Team1vsTeam2.reset_index()
WinLose = WinLose_19to23.reset_index()
Team1vsTeam2['IsWin'] = WinLose['IsTeam1Won']
Team1vsTeam2
```

```python
# export the after-clean dataset as the secondary analysis for the project
Team1vsTeam2.to_excel(r'Team1vsTeam2_2019-2023.xlsx', sheet_name='Sheet1', index = False) 
# Save 2019-2022 as train set and 2023 as test set or Save all together and use scikit-learn to split data into train and test set.
```
